# Handle NA values

```python
import pandas as pd

df = pd.read_csv('/content/weather_data.csv', parse_dates=['day']) # parse_dates will convert a column for "date"
df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace = True)
df
```

```python
# filling up null values with 0
df.fillna(0, inplace=True)
df
```

```python
# Filling up with 0 is not a good idea, therefore mean value or custom value is better
df.fillna({
    'temperature': df['temperature'].mean(),
    'windspeed': df['windspeed'].mean(),
    'event': 'No event'
}, inplace=True)

df
```

```python
# 'ffill' is forward fill, it will fill from the exact previous value
df.fillna(method='ffill')
```

```python
# backward fill
df.fillna(method='bfill')
```

```python
# Linear interpolation: fills gaps by connecting known points with a straight line.
df.interpolate()
```

```python
# in order to drop NA rows
# by default it will drop when it gets any value is empty
df.dropna()
```

```python
# now, it will drop only that rows where row is totally empty
df.dropna(how='all')
```

```python
# now, it will drop only that rows, where it will get 2 NA values
df.dropna(thresh=2)
```

```python
import pandas as pd
df = pd.read_csv('/content/weather_data.csv')
df
```

```python
# replacing -99999 and -88888 with NaN
import numpy as np
new_df = df.replace([-99999,-88888], np.nan)
new_df
```

```python
# in order to replace one by one
df.replace({
    'temperature' : -99999,
    'windspeed' : [-99999, -88888],
    'event' : 'no event'
}, np.nan)

df
```

```python
# alternative
df.replace({
    -99999 : np.nan,
    -88888 : np.nan,
    'no event' : 'Sunny'
})

df
```

### <font color="purple">**Question3**</font>
Fill the null values in respective columns like below:

# Concat and Merge

```python
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather], keys = ['india', 'us'])
df
```

```python
# in order to show values of 'india'
df.loc['india']
```

```python
# custom dataframe 1 for concatinate
temperature_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "temperature": [32, 45, 30]
}, index=[0,1,2])

temperature_df
```

```python
# custom dataframe 2 for concatinate
windspeed_df = pd.DataFrame({
    "city": ["delhi", "mumbai"],
    "windspeed": [7, 12]
}, index=[1,0])

windspeed_df
```

```python
# concatinating
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
# custom dataframe 1 for merge
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35]
})
df1
```

```python
# custom dataframe 2 for merge
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "orlando"],
    "humidity": [68, 65, 75]
})
df2
```

```python
# merge
pd.merge(df1, df2, on='city')
```

```python
# dataframe 1 for "inner join", "outer join", "left join", "right join"
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 38]
})
df1
```

```python
# dataframe 2 for "inner join", "outer join", "left join", "right join"
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "san francisco"],
    "humidity": [68, 65, 75]
})
df2
```

```python
# inner join
pd.merge(df1, df2, on='city', how='inner')
```

```python
# left join
pd.merge(df1, df2, on='city', how='left')
```

```python
# right join
pd.merge(df1, df2, on='city', how='right')
```

```python
# outer join
pd.merge(df1, df2, on='city', how='outer')
```
